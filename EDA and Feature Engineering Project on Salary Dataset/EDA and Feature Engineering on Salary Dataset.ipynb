```python
# Outliers Detection and Removing Outliers.

# Detect Outliers Using IQR.
Q1 = df['Salaries Reported'].quantile(0.25)  # 25th percentile
Q3 = df['Salaries Reported'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Any value below lower_bound or above upper_bound is considered an Outlier.

# Remove Outliers.

# Filter the dataset to remove Outliers.

df_clean = df[(df['Salaries Reported'] >= lower_bound) & (df['Salaries Reported'] <= upper_bound)]

print("Original dataset size:", df.shape)
print("Cleaned dataset size:", df_clean.shape)

# df_clean → dataset without outliers in battery_power.

# You can now use df_clean for analysis or modeling.


import seaborn as sns
import matplotlib.pyplot as plt

# Before removing outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Salary'])
plt.title("Salaries Reported - Before Removing Outliers")
plt.show()

# After removing outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df_clean['Salary'])
plt.title("Salaries reported - After Removing Outliers")
plt.show()


# From the Output, the Observations which can be Recorded are: 
#1) Q1 (25th percentile) = 1.

#2) Q3 (75th percentile) = 1.

#3) So, IQR = Q3 – Q1 = 1 – 1 = 0.

#4) Meaning, only values equal to 1 are considered “normal”, and everything else (2, 3, 4, …) is treated as an Outlier.

#5) This kept only rows where Salaries Reported = 1, removing all others.

#1) Original dataset size = 22,770 rows

#2) Cleaned dataset size = 18,206 rows

#3) This means 4,564 rows were dropped (because Salaries Reported was not equal to 1).

#6) Salaries Reported – Before Removing Outliers.

#1) The boxplot is very compressed around 1.

#2) This shows most values of Salaries Reported are equal to 1, while some higher values stretch far beyond (outliers).


#7) Salaries Reported – After Removing Outliers

#1) After applying the IQR method, only rows with Salaries Reported = 1 remained.

#2) The boxplot now shows a single line at 1 with no variation.

#3) This is because Q1 = Q3 = 1 → IQR = 0, so everything except 1 was removed.


# It is Clear from the Box - Plot , there are  Outliers in Salaries Reported Column.
```

## Feature Engineering on this Dataset. 

```python
#) Scaling / Normalization.

#1) Scaling (or Normalization) is a way of resizing your numbers so that all your features (columns) are on a similar scale.

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Salary_scaled'] = scaler.fit_transform(df[['Salary']])
print(df['Salary_scaled'])
```

```python
# Binning / Discretization.

# Convert a Numerical Variable into Categories(bins). Useful for features like age, battery, price.

# Binning battery_power into 3 categories: Low, Medium, High.

df['Salary_bin'] = pd.cut(df['Salary'], bins=2, labels=['Low','High'])
print(df['Salary_bin'])
```

```python
# Label Encoding.

#1) Label Encoding is a feature engineering technique used to convert categorical values (text data) into numeric labels.

#2) Each unique category in a column is assigned an integer value.

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['employment_status_le'] = le.fit_transform(df['Employment Status'])
print(df['employment_status_le'])

#1) 1 - Employed 
#2) 0 - Not Employed
```
